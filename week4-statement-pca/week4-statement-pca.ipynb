```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

# 1. Загрузите данные close_prices.csv. 
В этом файле приведены цены акций 30 компаний на закрытии торгов за каждый день периода. 

```python
data = pd.read_csv('close_prices.csv')
```

```python
data
```

```python
index = data.loc[:,'AXP':]
```

# 2. На загруженных данных обучите преобразование PCA с числом компоненты равным 10. 
Скольких компонент хватит, чтобы объяснить 90% дисперсии?

```python
pca = PCA(n_components=10)
pca.fit(index)

for i in range(30):
    pca = PCA(n_components=i)
    pca.fit(index)
    if sum(pca.explained_variance_ratio_) >= 0.9:
        i_max = i
        break
print(i_max)

f = open('answer_1.txt', 'w')
f.write(str(i_max))
f.close()
```

# 3. Примените построенное преобразование к исходным данным и возьмите значения первой компоненты. 

```python
X = pd.DataFrame(pca.transform(index))[0]
```

# 4. Загрузите информацию об индексе Доу-Джонса из файла djia_index.csv. 
Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса? 

```python
doy = pd.read_csv('djia_index.csv')['^DJI']
```

```python
Pirs = np.corrcoef(X, doy)
f = open('answer_2.txt', 'w')
f.write(str(np.round(Pirs[0, 1], 2)))
f.close()
```

# 5. Загрузите информацию об индексе Доу-Джонса из файла djia_index.csv. 
Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса? 

```python
f = open('answer_3.txt', 'w')
f.write(index.columns[np.argmax(pca.components_[0])])
f.close()
```
